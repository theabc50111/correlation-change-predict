```python
import pandas as pd
```

# Comparison ARIMA model & ARIMA prediciton

```python
ori_arima_pred_df = pd.read_csv('../../../ARIMA-LSTM-hybrid-corrcoef-predict/YWT_data_generation/new_asset_after_arima_pred_paper_eva_5.csv')
ywt_arima_pred_df = pd.read_csv('../exploration_hybrid_model_prediction/stock_correlation_prediction/use_hybrid_model/paper_eva_5_res.csv')

ori_arima_pred_df = ori_arima_pred_df.iloc[3::4,:]
ori_arima_pred_df['portfolio'] = ori_arima_pred_df['portfolio'].apply(lambda x:x[:-6])
ori_arima_pred_df.rename(columns = {'0':'arima_pred'}, inplace = True)
ywt_arima_pred_df = ywt_arima_pred_df.loc[::, ['portfolio', 'ground_truth', 'arima_pred', 'arima_model']]

compar_arima_pred_df = pd.merge(ywt_arima_pred_df, ori_arima_pred_df, on=['portfolio'], suffixes=['_ywt', '_ori'])
compar_arima_pred_df['arima_pred_diff'] = compar_arima_pred_df['arima_pred_ori'] - compar_arima_pred_df['arima_pred_ywt']
mask_arima_model = compar_arima_pred_df['arima_model_ori'] == compar_arima_pred_df['arima_model_ywt']
compar_arima_pred_df['arima_model_match'] = mask_arima_model

display(compar_arima_pred_df.iloc[(compar_arima_pred_df['arima_model_match']==False).tolist(), :])
display(compar_arima_pred_df.iloc[compar_arima_pred_df['arima_model_match'].tolist(), :])
```

# Comparison ARIMA residual

```python
ori_arima_resid_df = pd.read_csv('../../../ARIMA-LSTM-hybrid-corrcoef-predict/YWT_data_generation/new_asset_after_arima_resdiual_paper_eva_5.csv')
ywt_arima_resid_df = pd.read_csv('../exploration_hybrid_model_prediction/stock_correlation_prediction/use_hybrid_model/paper_eva_5_res_arima_resid.csv')

ori_arima_resid_df = ori_arima_resid_df.iloc[3::4, :21]
compar_arima_resid_df = pd.concat([ori_arima_resid_df, ywt_arima_resid_df]).sort_values('portfolio')
arima_resid_diff_df = compar_arima_resid_df.iloc[::2, 1::].reset_index(drop=True) - compar_arima_resid_df.iloc[1::2, 1::].reset_index(drop=True)
arima_resid_diff_df.index = compar_arima_resid_df['portfolio'][::2]

# display(ori_arima_resid_df.head(), ywt_arima_resid_df.head())
display(compar_arima_resid_df.head(10))
display(arima_resid_diff_df.head())
display(arima_resid_diff_df.abs().sum(axis=1).sort_values(ascending=False))
```
